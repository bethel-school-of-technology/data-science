# Work Cycle of a Data Scientist

The typical work cycle of a data scientist goes something like this:

<b>PROBLEM STATEMENT</b><br>
- Talk to the process owners and try to gain some clarity on the problem: Specifically get clear on what action will be taken after your insights are delivered.<p>

<b>DATA ACQUISITION</b><br> - Get an idea of where the data comes from, how it is generated, are there any knobs or dials that can be turned to change the outcome, etc.<p>
    
<b>DATA WRANGLING</b><br>  - Explore, clean up, and wrangle the data to better understand it. <p>

<b>EXPLORATORY DATA ANALYSIS</b><br>-   This can include performing some simple visualizations to better understand the data and determining which variables make sense as independent versus dependent variables.<p>
<b>PRE-PROCESSING</b><br> <p>
<b>MODELING</b><br> <p>
<b>INSIGHT DELIVERY</b><br> <p>

After that, it is wash, rinse, repeat.  In theory, projects like this could go on for the rest of your career, but in all likelihood, your company will not want to lose you to a single data set for the rest of time.  The big exception to this is if you are doing research - in this case, you could literally spend 20 years or more focusing on a single data set.

Statistical analysis is an iterative process.  It is very unusual for someone to bring you a pristine data set and ask for one simple analysis. The reality is that you usually have some freedom to explore the data, and go where your nose leads you.  You might end up going down a bunch of rabbit holes that produce no results, but this, too, is part of the job. As a data scientist, you have a chance to utilize both your creativity and your curiosity.

## Sources of Data

You can basically choose data from anywhere, as long as it is freely available and you have permission to use it.  Consider the following data source options:

* Collecting your own data
* Asking your employer for data to help solve his or her work problems
* Asking a family member or friend for data to help solve their problems
* Using readily available data from the internet

There are pros and cons to all of these options. 

---

### Collecting Your Own Data

If you feel strongly about collecting your own data, you will need to undergo a review to ensure that your data collection will not cause harm to others.  It will also take a lot of extra time to create a survey, distribute that survey, and receive results upon that survey, so make sure you are really excited about demonstrating surveying techniques and that the data does not already exist elsewhere.  Your analyses will most likely be less involved, due to the time sink of data collection.

---

### Asking Others for their Data

Whether you are getting data from an employer, a friend, or a family member, you will need their explicit permission to utilize that data.  This can also take time, especially if you're using data from your company - your request for data may need to work its way up the chain of command.  It may also put limits on how you can use and present upon that data.  Someone may require you to anonymize the data or to not record your presentation. However, using "real-world" data that actually solves an actual need can give you a leg up in the employment world, and if you are using data from your company, you can list it under your job description on your resume, and thus increases your chances of employment.

## Considerations when Picking Data

There are many things to consider when picking a dataset.  

* **Do you like data wrangling?** Some datasets are already nice and clean, and ready for analysis.  You just need to pop 'em in a statistical analysis or machine learning algorithm and you are good to go.  Others, especially those found on data.gov, may require a lot of wrangling.  How much are you willing to do? Keep in mind that data wrangling is often 90% of real projects and that employers will expect you to have some wrangling skills.

* **Do you have enough data?** Most of the more complicated analyses require at least 200 rows as a rough guideline.

* **Do you have too much data?** Given that you don't actually have a big data cluster running multiple nodes, and you probably don't want to pay for one, choose a dataset that is manageable in size so that you can run it yourself.

* **How much data is missing?** If you find that 75% of the data is missing, then you may end up not having enogh data after all.  And if that data is essentatial to the topic at hand, then that's another reason you might not be able to use it.

* **Is there data on your topic?** Again, if you can't find any data on a particular topic, it may be better to pick your second favorite topic that go on a wild goose chase.

* **Will you have to piece together data from different sources?** Doing so can be awesome and fun, and demonstrates a wider range of skills.  But it can also be a nightmare! If you are interested in doing some data joining, then make sure you pick data that has a common key and that makes sense to combine.

* **What stipulations, if any, are put on the use of the data?** Do you need to pay a fee? Report on your findings somewhere? Make everything anonymous? Add in a reference to the data source? It's a good idea to find out in advance so you are not surprised later.

* **Is the data commonly used?** You want to demonstrate to employers that you can work off the beaten track.  If you pick a very common dataset that has lots of tutorials about how to analyze it, then you run the chance of being "same old, same old" and not drawing attention to yourself, no matter how well you completed the project. Especially avoid datasets built into Python and R, since those are some of the most overused datasets out there!

Paying attention to some of these considerations can really make or break you - thinking through the ramifications of your dataset in advance can make the difference between a project in which you spend all of your time cursing your data and a project that is a breeze and moves along smoothly. Snooping very briefly through the data and considering the above questions will save you a lot of time later.

# Choosing your Evaluation Questions

Now that you have your data picked out, its time to figure out what questions you want to answer with that data! 

---

## What is an Evaluation Question?

Yes, you're asking questions about questions now! So meta! An *evaluation question* is an intelligent question you can ask about your data in order to help out your company.  These questions will be the guiding principles that help organize your data exploration.  They can be relatively general, as long as you can translate them into specifics.  For instance, you might ask a question like "How do health metrics differ between ethnicities?" Although that is somewhat general, as "health metrics" and "ethnicities" can encompass many different components and variables, you can use your data to point towards specific components there.  If you have columns in your data about disability level, number of times someone has gone to the doctor in the last month, and the amount of money they have spent on medications, you have now quantified, or translated, "health metrics" into something that you can actually look at and measure. 

You should have at least two evaluation questions for your final project, and you should most likely be able to use several different data pieces in order to answer them.

# Choosing your Analyses

Once you have at least two evaluation questions, your next step is to determine what data and analyses you can use to actually answer those questions. 

First, start by identifying your independent, or predictor, variable, (X) and your dependent, or outcome, variable (y).  Remember that your dependent variable is what you are trying to predict - it is the thing you want to know about.  Your independent variable or variables (most advances analyses will allow you to have more than one) are the things that you think might influence the dependent variable. 

Second, you will need to identify what data types the independent and dependent variables are.  Do you have categorical variables, that are just groupings? Or do you have something continuos, that is a number, and when you have a higher number, it means that you have more of the variable? Don't be fooled - you can have a categorical variable that looks like a number, but isn't actually continuous.  For instance, what if you recode the color of dogs from brown, black, and white to 0, 1, and 2? It may look like a number, but does having a 1 mean that there is more fur or more dog, versus having a zero? No! So even though the variable has numbers, it is categorical, not continuous.

## Executive Summary

The executive summary is a quick review of **everything** that is in the rest of your document. Especially if you have a military background, you may have heard this also referred to as a *"BLUF" - bottom line up front*. It's a way for the important people in your company to get the down and dirty details quickly and without a lot of extra reading. 

Even though it comes at the beginning, sometimes it can be easier to write the executive summary at the end - it's easier to summarize all the details once you already know them!

---

## Business Objectives

Your business objectives should include your evaluation questions. What are you trying to answer or solve with your data work?

---

## Background

You can't assume that everyone has the same level of subject matter expertise on a particular topic, so in the background section, you want to give an overview of the topic, the problem, and why it matters. Think about this in terms of building a business case: if you were to ask for funding to do this project, what would convince a client, manager, or other stakeholder to give it the green light? You want to include that information here.

---

## Scope

Imagine your project is on a stage, and it can only be followed by one spotlight. What are the components that the spotlight will hit? Those factors are considered *in scope*. Anything that is relevant to your project and you're willing to tackle will be included, but anything that is a "rabbit trail" leading you out of the spotlight or off the stage altogether will be excluded, which is called *out of scope*. It's very important to define your scope thoroughly because if you don't, its easy for other people to shift you off course. Getting derailed by others is called *scope creep*, and it can be deadly to delivering a project on-time and on-target. When other people suddenly start asking you for additional other components, and you haven't budgeted for those other needs, resources, and time, your project can get off track quickly. Sometimes this happens because of genuine interest or enthusiasm on the part of others, and sometimes it can happen because they have their own political agenda within the organization.

The stalwart guardian against scope creep is a well-defined scope! What are you willing to do for this project, and what are you unwilling to do? Think about things like:

* What programs are you willing to use and what are you unwilling to use?
* Are you willing to learn any new skills for this?
* Are you willing to add or change data if asked?
* What analyses are you willing to complete and which ones are you not?

While your instructor will not throw you out of scope, really thinking through this section will help you tremendously in the business world.

## Assumptions

Assumptions are outside factors that the project depends upon, but which you have little to no control over. For instance, you may assume a program such as Python will be working and not buggy. It's important to note the things you are depending upon, because their failure could severely derail you project. A good project manager is able to navigate around failed assumptions if they happen.

---

## Limitations

A limitation is a threat to your project that is internal to you and your team. It's a known weakness that you want to document so that you can plan around it. Remember that saying, "plan for the worst and hope for the best?" That is exactly what you are doing here, both in this limitations section and also in this proposal document as a whole. Limitations could be known gaps in your knowledge or skills, or they could be things like scheduling, communication, and timing issues. It's a great idea to discuss limitations with your team, because again, it allows you to plan for them instead of letting them derail you.

---

## Risks

A risk is similar to a limitation in that it is a threat to your project, but instead of being internal, a risk is external to your project or team. Risks are things that you don't necessarily have control over. This makes them a little more difficult to think through and plan, but the more you do so, the more stable your project will be if you hit some waves. 

A great example of a risk is loss of internet due to bad weather. What will you or your teammates do if you lose internet? You can't control that, but you can plan ahead to deal with it. Maybe you can head to a family or friend's house. Maybe you can hit a public library or a cafe with internet.

Your goal is to put on your pessimism hat and think through all the weird and wooly scenarios, and then put an action plan in place with which to deal with them!
